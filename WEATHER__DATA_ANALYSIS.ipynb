```python
#  NUMBER OF DAYS WEATHER IS CLEAR
clear_count=df['Weather'].value_counts().get('Clear', 0)
print("The number of days the weather was clear is :",clear_count)
```

```python
# MOST OF THE TIME THE WEATHER CONDITION
most_of_time=df['Weather'].mode()
print("Most of the time the weather is: ",most_of_time)
```

```python
# AVERAGE VALUES OF VISIBILITY, TEMPRATURE AND WIND SPEED
print("Mean visibility_km is:",df['Visibility_km'].mean())
print("Mean temprature is:",df['Temp_C'].mean())
print("Mean wind spee is:",df['Wind Speed_km/h'].mean())
```
